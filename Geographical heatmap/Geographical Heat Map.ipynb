```python
import pandas as pd
from folium import Map
from folium.plugins import HeatMap
```

```python
for_map = pd.read_csv('data.csv')
```

```python
for_map
```

```python
hmap = Map(location=[30.272081, -97.754014], zoom_start=15, )
```

```python
hm_wide = HeatMap(
    list(zip(for_map.latitude.values, for_map.longitude.values)),
    min_opacity=0.2,
    radius=17, 
    blur=15, 
    max_zoom=1,
)
```

```python
hmap.add_child(hm_wide)
```
