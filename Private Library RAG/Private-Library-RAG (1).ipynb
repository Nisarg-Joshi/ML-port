```python
import os
import bs4
# Suppress the User Agent warning
os.environ["USER_AGENT"] = "MyRAGApp/1.0"

from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore

print("1. Loading document...")
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs={"parse_only": bs4.SoupStrainer(class_=("post-title", "post-header", "post-content"))},
)
docs = loader.load()

print("2. Splitting text...")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
all_splits = text_splitter.split_documents(docs)

print("3. Initializing Embeddings (NumPy check passed)...")
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

print("4. Indexing into Vector Store...")
vector_store = InMemoryVectorStore(embeddings)
vector_store.add_documents(documents=all_splits)

print("-" * 30)
print("SUCCESS! Your RAG system is ready.")
```

```python
# 1. Define your question
query = "What are the main components of an LLM-powered agent system?"

# 2. Search the vector store for the most relevant parts of the blog
docs = vector_store.similarity_search(query, k=3)

# 3. Print the results
print(f"--- Question: {query} ---\n")
for i, doc in enumerate(docs):
    print(f"Result {i+1}:")
    print(doc.page_content[:300] + "...") # Print first 300 characters of each match
    print("-" * 20)
```

```python
import os
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_community.llms import GPT4All
from langchain.chains import RetrievalQA

# 1. Re-initialize the pieces (needed after a restart)
print("Setting up the engine...")
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_store = InMemoryVectorStore(embeddings)
vector_store.add_documents(documents=all_splits) # Use the splits we made earlier

# 2. Load the Local AI Model 
# This will download the ~3GB model file to your computer the first time.
print("Loading Local LLM (Orca Mini)...")
llm = GPT4All(model="orca-mini-3b-gguf2-q4_0.gguf")

# 3. Create the RetrievalQA Chain
# This is the "Glue" that makes it RAG.
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)

# 4. Ask the final question
question = "What are the core components of an LLM-powered agent system?"
print(f"\nAsking the AI: {question}")

response = qa_chain.invoke(question)

print("\n" + "="*30)
print("FINAL RAG ANSWER:")
print(response["result"])
print("="*30)
```

```python
import os
import bs4
# Suppress the User Agent warning
os.environ["USER_AGENT"] = "MyRAGApp/1.0"

from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_community.llms import GPT4All
from langchain.chains import RetrievalQA

print("1. Scraping the blog post...")
loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs={"parse_only": bs4.SoupStrainer(class_=("post-title", "post-header", "post-content"))},
)
docs = loader.load()

print("2. Splitting into chunks...")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
all_splits = text_splitter.split_documents(docs)

print("3. Loading Free Embeddings (Local)...")
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_store = InMemoryVectorStore(embeddings)
vector_store.add_documents(documents=all_splits)

print("4. Loading Local LLM (this may take a minute)...")
# Note: If it hangs here, it's just downloading the 3GB model file.
llm = GPT4All(model="orca-mini-3b-gguf2-q4_0.gguf")

print("5. Connecting the RAG Pipeline...")
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vector_store.as_retriever()
)

# The final question
question = "What are the main components of an LLM-powered agent system?"
print(f"\nAsking the AI: {question}")

response = qa_chain.invoke(question)

print("\n" + "="*30)
print("FINAL RAG ANSWER:")
print(response["result"])
print("="*30)
```
